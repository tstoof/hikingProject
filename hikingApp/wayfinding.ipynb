```python
import openrouteservice
from openrouteservice.convert import decode_polyline

# Initialize the client with your API key
client = openrouteservice.Client(key="5b3ce3597851110001cf6248b2b60bf1864a44568d06951c3b3b47a2")

# Define coordinates for the start and end points [longitude, latitude]
start = [6.3326, 52.7594]
end = [6.3415, 52.8212]   

# Request a route
route = client.directions(
    coordinates=[start, end],
    profile='foot-hiking',  # Other options: 'cycling-regular', 'foot-walking', etc.
    # format='geojson'        # Response format
)
route = decode_polyline(route['routes'][0]['geometry'])
print(route)
# Print the route geometry
# print(route['features'][0]['geometry'])
```

```python
import folium

# Extract geometry from the route
coordinates = route['features'][0]['geometry']['coordinates']

# Create a map centered on the start point
map_ = folium.Map(location=[start[1], start[0]], zoom_start=14)

# Add the route to the map
folium.PolyLine([(lat, lon) for lon, lat in coordinates], color="blue", weight=5).add_to(map_)

# Display the map
map_.save("route_map.html")
print("Map saved as 'route_map.html'")
```
